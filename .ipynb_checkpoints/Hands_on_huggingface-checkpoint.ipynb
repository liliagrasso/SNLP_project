```python
import transformers
import pandas as pd
```

```python
pandas_dataset = pd.read_csv("non_tokenized.csv")
print(pandas_dataset.columns)
```

```python
pandas_dataset.dropna(inplace=True)
```

```python
pandas_dataset=pandas_dataset.groupby(["title","side","ideology","party","gender","ethnicity","religion"],as_index=False)["text"].apply(lambda x: ' '.join(x))
```

```python
#splitting into train set
dataset_train = pandas_dataset[:int(0.8*len(pandas_dataset))]
```

```python
tokenizer = transformers.AutoTokenizer.from_pretrained("bert-base-uncased")
model = transformers.AutoModelForSequenceClassification.from_pretrained("bert-base-cased")

feature_extractor = transformers.pipeline('feature-extraction', model=model, tokenizer= tokenizer)
dataset_train["title"].apply(feature_extractor)
```

```python
tensorBatchEncodingTrainX = tokenizer(text=feature_extractor(dataset_train["title"]),
                                      text_pair=dataset_train["side"],
                                      add_special_tokens=True,
                                      truncation="longest_first",
                                      padding="max_length",
                                      #max_length=None if max_length_x <= 0 else max_length_x,
                                      return_tensors="pt")
```
